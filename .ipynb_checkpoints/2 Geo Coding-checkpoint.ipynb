## 2. Geocoding through Google API

```python
import requests
```

```python
location = [ lli for lli in pd1_sub["Geographic Location"].values.tolist()  ]
```

```python
#Google (Note: I have deleted the api key)
location1 =list(set(location))
location_geo =[]
location_des = []
lon=[]
lat=[]
for i in range(len(location1)):
    loi = location1[i]
    test_geo =loi
    test_geo= test_geo.replace(" ","+")
    key =""
    url ="https://maps.googleapis.com/maps/api/geocode/json?address="+test_geo+"&key="+key
    response = requests.get(url)
    re1 = response.json()
    try:
        location_geo.append(re1["results"][0]["geometry"]['location'])
        location_des.append(re1["results"][0])
        lon.append(re1["results"][0]["geometry"]['location']["lng"])
        lat.append(re1["results"][0]["geometry"]['location']["lat"])
        print(i)
    except:
        location_geo.append("")
        location_des.append("")
        lon.append("")
        lat.append("")
        print("wrong....................."+str(i))
```

```python
ad1=[]
for i in range(len(location_des)):
    if location_des[i]!="":
        fullname=""

        delocation= location_des[i]["address_components"]
        for j in range(len(delocation)):
            fullname = fullname+ delocation[j]['long_name']+","
        fullname= fullname[:-1]
        ad1.append(fullname)
    else:
        ad1.append("")
ad1
```

```python
c = {
    "Geographic Location":location1,
    "location":location_geo,
    "lon":lon,
    "lat":lat,
    "location_fromgoogle":ad1,
    "Description":location_des
}
geolocation = pd.DataFrame(c)
```
